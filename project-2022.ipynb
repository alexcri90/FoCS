### 3. For each (primary) breed, determine the ratio between the number of dogs of `Mixed Breed` and those not of Mixed Breed. Hint: look at the `secondary_breed`.

```python
total_number_and_mixed_breed = dogs.groupby('breed_primary')['breed_mixed'].agg({'count','sum'}).rename(columns= {'count':'total_number', 'sum': 'breed_mixed_number'})
total_number_and_mixed_breed
```

```python
#Define the function compute_ratio, which compute the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed
def compute_ratio(row):
    
    total_number = row['total_number']
    breed_mixed_number = row['breed_mixed_number']
    if total_number == breed_mixed_number:
        return 'NaN'
    return breed_mixed_number/(total_number - breed_mixed_number)
```

```python
#apply the function compute_ratio to the current dataset.
total_number_and_mixed_breed['ratio']= total_number_and_mixed_breed.apply(compute_ratio, axis = 1)
total_number_and_mixed_breed.head()
```

We check what values the affected columns can take, we create a function instead of 'unique' method because it return values follow by ',' and it will be easy to copy for replace

```python
def unique(lista):
  uniq = []
  for x in lista:
      if x not in uniq:
          uniq.append(x)
  return uniq
```
